#Analysis Part:


##Sales Analysis Based on Time

###Calculating Monthly Sales:

```python
sales_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month
```

###Monthly Sales in Descending Order

```python
# Sort by Sales in descending order
sales_by_month_des = sales_by_month.sort_values(by='Sales', ascending=False)
sales_by_month_des
```

###Month with Max and Min Sales

```python
# Month with maximum sales
max_month = sales_by_month['Sales'].idxmax()
max_sales = sales_by_month['Sales'].max()

# Month with minimum sales
min_month = sales_by_month['Sales'].idxmin()
min_sales = sales_by_month['Sales'].min()

print(f"Highest Sales Month: {max_month} - Sales: {max_sales}")
print(f"Lowest Sales Month: {min_month} - Sales: {min_sales}")
```

###Day with Highest and Lowest Sales

```python
# Assuming 'df' is your original DataFrame and has 'Order Date' and 'Sales' columns

# Make sure 'Order Date' is in datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by date and sum the sales for each day
daily_sales = df.groupby('Order Date')['Sales'].sum()

# Find the day with the highest and lowest sales
max_day = daily_sales.idxmax()
max_sales = daily_sales.max()

min_day = daily_sales.idxmin()
min_sales = daily_sales.min()

print(f"Day with Highest Sales: {max_day.date()} - Sales: {max_sales}")
print(f"Day with Lowest Sales: {min_day.date()} - Sales: {min_sales}")
```

###Visualizing Monthly Sales Trends

```python
import plotly.express as px

fig = px.line(sales_by_month, x=sales_by_month.index, y='Sales', title='Sales')
fig.update_layout(
    width=800,
    height=400,
    showlegend=False,
    plot_bgcolor='white',
    xaxis=dict(showline=True, linewidth=1, linecolor='black'),
    yaxis=dict(showline=True, linewidth=1, linecolor='black')
)
fig.show()
```

###MOM Sales Trend

```python


# Step 1: Ensure 'Order Date' is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Step 2: Create Year-Month column
df['YearMonth'] = df['Order Date'].dt.to_period('M').astype(str)

# Step 3: Group by Year-Month and sum Sales
monthly_sales = df.groupby('YearMonth')['Sales'].sum().reset_index()

# Step 4: Calculate MoM % change
monthly_sales['MoM Change %'] = monthly_sales['Sales'].pct_change() * 100

# Step 5: Plot
fig = px.line(monthly_sales,
              x='YearMonth',
              y='Sales',
              title='Month-over-Month Sales Trend',
              markers=True)

# Optional: Add bar chart for MoM change
fig.add_bar(x=monthly_sales['YearMonth'],
            y=monthly_sales['MoM Change %'],
            name='MoM Change %',
            marker_color='orange',
            yaxis='y2')

# Update layout for dual y-axes
fig.update_layout(
    yaxis=dict(title='Sales'),
    yaxis2=dict(title='MoM Change %', overlaying='y', side='right'),
    title_font=dict(size=22),
    xaxis_title='Month',
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```

###YOY Sales Trend

```python
import pandas as pd
import plotly.express as px

# Step 1: Ensure 'Order Date' is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Step 2: Extract Year
df['Year'] = df['Order Date'].dt.year

# Step 3: Group by Year and sum Sales
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

# Step 4: Calculate YoY % change
yearly_sales['YoY Change %'] = yearly_sales['Sales'].pct_change() * 100

# Step 5: Plot
fig = px.bar(yearly_sales,
             x='Year',
             y='Sales',
             title='Year-over-Year Sales with YoY % Change',
             text='Sales')

# Add line for YoY % change
fig.add_scatter(x=yearly_sales['Year'],
                y=yearly_sales['YoY Change %'],
                mode='lines+markers',
                name='YoY Change %',
                yaxis='y2',
                line=dict(color='orange', dash='dash'))

# Update layout for dual y-axes
fig.update_layout(
    yaxis=dict(title='Sales'),
    yaxis2=dict(title='YoY Change %', overlaying='y', side='right'),
    title_font=dict(size=22),
    xaxis_title='Year',
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```

```python
import plotly.express as px

# Step 1: Group and get best/worst
sales_by_category = df.groupby('Category')['Sales'].sum()
best_cat = sales_by_category.idxmax()
worst_cat = sales_by_category.idxmin()

# Step 2: Convert to DataFrame
sales_df = sales_by_category.reset_index()
sales_df.columns = ['Category', 'Sales']

# Step 3: Define custom color sequence
colors = []
for cat in sales_df['Category']:
    if cat == best_cat:
        colors.append('green')   # Best - Green
    elif cat == worst_cat:
        colors.append('red')     # Worst - Red
    else:
        colors.append('lightgray')  # Others - Gray

# Step 4: Plot
fig = px.pie(sales_df,
             values='Sales',
             names='Category',
             hole=0.5,
             color_discrete_sequence=colors)

# Step 5: Styling
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(
    title_text='Sales Analysis by Category (Best & Worst Highlighted)',
    title_font=dict(size=22)
)

fig.show()
```

```python
sales_by_customer_segment = df.groupby('Segment')['Sales'].sum().sort_values(ascending=True)
profit_by_customer_segment = df.groupby('Segment')['Profit'].sum().sort_values(ascending=True)
print(sales_by_customer_segment)
print(profit_by_customer_segment)
import pandas as pd
import plotly.express as px

import pandas as pd
import plotly.express as px

# Step 1: Group and sort
sales_by_customer_segment = df.groupby('Segment')['Sales'].sum().sort_values(ascending=True)
profit_by_customer_segment = df.groupby('Segment')['Profit'].sum().sort_values(ascending=True)

# Step 2: Combine into one DataFrame
combined = pd.DataFrame({
    'Sales': sales_by_customer_segment,
    'Profit': profit_by_customer_segment
}).reset_index()

# Step 3: Melt for grouped bar chart
melted = combined.melt(id_vars='Segment', value_vars=['Sales', 'Profit'],
                       var_name='Metric', value_name='Amount')

# Step 4: Plot grouped bar chart
fig = px.bar(melted,
             x='Segment',
             y='Amount',
             color='Metric',
             barmode='group',
             title='Sales vs Profit by Customer Segment',
             text='Amount',
             color_discrete_map={'Sales': 'royalblue', 'Profit': 'seagreen'})

# Step 5: Styling
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(
    xaxis_title='Customer Segment',
    yaxis_title='Amount',
    title_font=dict(size=22)
)

fig.show()
```

##Sales to Profit Ratio:

A higher ratio means it takes more sales to generate profit (lower efficiency), and a lower ratio indicates better profitability.

```python
import pandas as pd

# Step 1: Group by Segment
sales_by_segment = df.groupby('Segment')['Sales'].sum()
profit_by_segment = df.groupby('Segment')['Profit'].sum()

# Step 2: Combine into DataFrame
ratio_df = pd.DataFrame({
    'Sales': sales_by_segment,
    'Profit': profit_by_segment
})

# Step 3: Calculate Sales-to-Profit Ratio
ratio_df['Sales-to-Profit Ratio'] = ratio_df['Sales'] / ratio_df['Profit']

# Optional: Format for easier viewing
ratio_df = ratio_df.reset_index()
print(ratio_df)
```
